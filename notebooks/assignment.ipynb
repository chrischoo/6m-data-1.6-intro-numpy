```python
# Question 3: Boolean masks: Work with targets

import numpy as np

names = np.array(["Ana", "Ben", "Chen", "Dana", "Eli", "Fatima", "George", "Hui"])
spend = np.array([200, 150, 300, 120, 180, 220, 160, 310])      # marketing spend
revenue = np.array([400, 180, 500, 100, 220, 260, 150, 600])   # revenue

roi = revenue / spend
print(roi, '\n')

roi_mask = roi >= 2.0
print('Mask for ROI >= 2.0: ', roi_mask, '\n')

# Apply mask to filter names, spend, and revenue for customers with ROI >= 2.0
print(names[roi_mask], '\n')
print(spend[roi_mask], '\n')
print(revenue[roi_mask], '\n')

roi_mask2 = spend >= 200
print ('Mask for Spend >= 200: ', roi_mask2, '\n')

roi_mask_combined = roi_mask & roi_mask2
print('Mask for Spend >=200 AND ROI >=2.0: ', roi_mask_combined, '\n')

print('Names for Combined ROI Mask: ', names[roi_mask_combined], '\n')

# This filtered group of customers has a high ROI and also a high marketing spend, which may indicate that they are valuable customers worth investing in further.
```

```python
# Question 4: Gradebook: Aggregations and broadcasting

import numpy as np
students = np.array(["S1", "S2", "S3", "S4", "S5"])
subjects = np.array(["Math", "Stats", "Python"])

scores = np.array([
    [75, 80, 85],
    [60, 65, 70],
    [90, 88, 92],
    [82, 79, 84],
    [70, 72, 78]
])

for i in range(len(students)):
    student_scores = scores[i]
    average_score = np.mean(student_scores)
    print(f"{students[i]}'s average score: {average_score:.2f}")

for i in range(len(subjects)):
    subject_scores = scores[:, i]
    average_score = np.mean(subject_scores)
    print(f"Average score for {subjects[i]}: {average_score:.2f}")

# 1. Calculate the mean for each subject (axis=0 averages down the columns)
subject_means = scores.mean(axis=0)
print ("Subject Means:", subject_means, '\n')

# 2. Subtract the means from the scores (Broadcasting)
scores_centered = scores - subject_means
print ("Centered Scores:\n", scores_centered, '\n')

raw_average = scores.mean(axis=1)
print ("Raw Averages:", raw_average, '\n')

centered_average = scores_centered.mean(axis=1)
print ("Centered Averages:", centered_average, '\n')

max_raw_average = np.max(raw_average)
print ("Max Raw Average:", max_raw_average, '\n')
student_with_max_raw_average = students[np.argmax(raw_average)]
print(f"Student with Max Raw Average: {student_with_max_raw_average} with a score of {max_raw_average:.2f}\n")

max_centered_average = np.max(centered_average)
print ("Max Centered Average:", max_centered_average, '\n')
student_with_max_centered_average = students[np.argmax(centered_average)]
print(f"Student with Max Centered Average: {student_with_max_centered_average} with a score of {max_centered_average:.2f}\n")

# The student with the highest raw average is S3, while the student with the highest centered average is also S3. This indicates that S3 performed well across all subjects, even after accounting for the average performance in each subject.
# Centered averages provide a more balanced view of student performance by normalizing for subject difficulty, while raw averages may be skewed by particularly high or low scores in certain subjects.
```

```python
# Question 5: Reshape and flatten: From daily to weekly

import numpy as np
daily_visits = np.array([
    120, 130, 125, 140, 150,
    160, 170, 155, 145, 135,
    128, 132, 138, 142, 148,
    152, 158, 162, 168, 172,
    180, 190, 185, 175, 165,
    155, 145, 135, 125, 115
])

daily_visits = daily_visits.reshape(6, 5)
print("Daily Visits:\n", daily_visits, '\n')

visits_per_week = daily_visits.sum(axis=1)
print("Total Visits Per Week:", visits_per_week, '\n')

average_visits_per_day = daily_visits.mean(axis=0)
print("Average Visits Per Day:", average_visits_per_day, '\n')

daily_visits = daily_visits.flatten()
print("Flattened Daily Visits:", daily_visits, '\n')

# Reshaping groups the data by 5 days per week, allowing us to analyze weekly trends.
# Summing across axis=1 gives us the total visits for each week, while averaging across axis=0 provides insights into daily patterns.
# Flattening the array allows us to treat all visits as a single sequence for further analysis or visualization.
```

```python
# Question 6: Mini-project: Simple scoring model

import numpy as np

# Each row: [page_views, time_on_site (minutes), past_purchases]
X = np.array([
    [10,  3.5,  0],
    [25,  5.0,  1],
    [40,  2.0,  0],
    [15, 10.0,  3],
    [30,  4.0,  2]
])

customers = np.array(["C1", "C2", "C3", "C4", "C5"])

# Define weights for each feature
weights = np.array([0.5, 0.3, 0.2])
print("Feature Weights:", weights, '\n')

scores_initial_weights = X.dot(weights)
print("Customer Scores:", scores_initial_weights, '\n')

weights_adjusted = np.array([0.5, 0.3, 3.0])
print("Adjusted Weights:", weights_adjusted, '\n')
scores_adjusted_weights = X.dot(weights_adjusted)
print("Customer Scores with Adjusted Weights:", scores_adjusted_weights, '\n')

# Customer with highest score using initial weights
max_score_initial = np.max(scores_initial_weights)
customer_with_max_score_initial = customers[np.argmax(scores_initial_weights)]
print(f"Customer with Highest Score (Initial Weights): {customer_with_max_score_initial} with a score of {max_score_initial:.2f}\n")

# Customer with highest score using adjusted weights
max_score_adjusted = np.max(scores_adjusted_weights)
customer_with_max_score_adjusted = customers[np.argmax(scores_adjusted_weights)]
print(f"Customer with Highest Score (Adjusted Weights): {customer_with_max_score_adjusted} with a score of {max_score_adjusted:.2f}\n")

# I prefer the adjusted weights as they place more emphasis on past purchases, which is a strong indicator of customer loyalty and potential for future purchases.
# This allows us to identify customers who have a history of purchasing, even if their page views and time on site are not as high, which can be valuable for targeted marketing efforts.
# However, I would prefer using time_on_site as the most important feature if I want to focus on customer engagement, especially if the site is relatively new and past purchase data is limited.
# Time on site can indicate interest and potential for conversion, even if the customer has not made a purchase yet.
```
