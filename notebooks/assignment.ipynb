```python
import numpy as np  
array_1d = np.arange(10, 20)
print(array_1d, '\n')
print('array_1d Shape:' , array_1d.shape , '\n')
print('array_1d NDimensions:' , array_1d.ndim , '\n')
print('array_1d Data Type:' , array_1d.dtype , '\n')

array_2d = np.full ([4,3], 1.5)
print(array_2d, '\n')
print('array_2d Shape:' , array_2d.shape , '\n')
print('array_2d NDimensions:' , array_2d.ndim , '\n')
print('array_2d Data Type:' , array_2d.dtype , '\n')

array_3d = np.zeros ([2,2,3])
print('array_3d Shape:' , array_3d.shape , '\n')
print('array_3d NDimensions:' , array_3d.ndim , '\n')
print('array_3d Data Type:' , array_3d.dtype , '\n')

array_2d = array_2d.astype('int')
print('array_2d Data Type:' , array_2d.dtype , '\n')

# I would use float data type instead of int when it contains decimal values such as prices and precise calculations. int only covers whole numbers and would not be suitable for such cases.
```
